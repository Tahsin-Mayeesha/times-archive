```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
archive = pd.read_csv("archive.csv")
```

```python
archive.head()
```

```python
archive["age of award"] = -(archive["Birth Year"]-archive["Year"])
```

## Most Popular Contexts : 

```python
contexts = archive['Context'].value_counts(ascending=True)
```

```python
contexts[-9:].plot(kind="barh",figsize=(7,7))
plt.title("Contexts appearing throughout history")
```

## Honor Designations :

```python
archive["Honor"].value_counts()
```

## Repeat Nominees : 

```python
names = archive["Name"].value_counts(ascending= True)[-14:].plot(kind="barh", figsize = (7,7), xticks=(0,1,2,3,4), grid = False)
plt.title("Recipients who have been featured twice or more")
```

## Age of the Recipients by Category : 

```python
archive["age of award"].hist(by=archive["Category"],figsize=(12,12));
```

## Categorical Distribution : 

```python
archive["Category"].value_counts(ascending=True).plot(kind="barh",xticks=(5,10,15,20,25,30,35,40,45))
```

```python
archive[archive["Category"]=="Science"]
```

```python
archive[archive["Category"]=="Philanthropy"]
```

```python
category = archive["Category"].value_counts(normalize=True,
                            ascending = True).plot(kind="barh",figsize = (7,7))
plt.title("Top  Categories for award getting")
plt.xlim()
plt.grid(False)

category.set_xticklabels(["{:1.1f}%".format(x*100) for x in category.get_xticks()])
plt.suptitle("Category-wise distribution of the award ",fontsize = 19)
plt.title("Politics and War are the dominant candidates")
;
```
